```python
# Check for null values

# Identify how many loans have number of borrowers equal to 1 and not null for borrower credit score
x1 = df_acquisition[['Num_Borrowers','Borrower_Credit_Score'
                     ,'Coborrower_Credit_Score']][(df_acquisition['Num_Borrowers'] == 1) 
                                                  & (df_acquisition['Coborrower_Credit_Score'].notnull())]
print('The count of the Num_Borrowers = 1 with values in Coborrower_Credit_Score column are: ' 
      + str(x1['Num_Borrowers'].count()))
      
# Identify how may loans have number of borrowers greater than 1 and NaN for co-borrower credit score
x2 = df_acquisition[(df_acquisition['Num_Borrowers'] > 1) & (df_acquisition['Coborrower_Credit_Score'].isna())]
print('The count of the Num_Borrowers > 1 and Coborrower_Credit_Score is NaN: ' 
      + str(x2['Num_Borrowers'].count()))
```

```python
# Drop rows with conflicting number of borrowers and co-borrower credit score data
#df_acquisition = df_acquisition.dropna(subset=['Coborrower_Credit_Score'], how='all')

df_acquisition = df_acquisition[((df_acquisition['Num_Borrowers'] > 1) 
                                 & (df_acquisition['Coborrower_Credit_Score'].notnull()) 
                                 | (df_acquisition['Num_Borrowers']==1) 
                                 & df_acquisition['Coborrower_Credit_Score'].isnull())]
```

```python
# Repeat check for null values
# Identify how many loans have number of borrowers equal to 1 and not null for borrower credit score
x1 = df_acquisition[['Num_Borrowers','Borrower_Credit_Score'
                     ,'Coborrower_Credit_Score']][(df_acquisition['Num_Borrowers'] == 1) 
                                                  & (df_acquisition['Coborrower_Credit_Score'].notnull())]
print('The count of the Num_Borrowers = 1 with values in Coborrower_Credit_Score column are: ' 
      + str(x1['Num_Borrowers'].count()))
      
# Identify how may loans have number of borrowers greater than 1 and NaN for co-borrower credit score
x2 = df_acquisition[(df_acquisition['Num_Borrowers'] > 1) & (df_acquisition['Coborrower_Credit_Score'].isna())]
print('The count of the Num_Borrowers > 1 and Coborrower_Credit_Score is NaN: ' 
      + str(x2['Num_Borrowers'].count()))
```
